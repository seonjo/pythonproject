```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
import numpy as np
```

```python
iris = load_iris()
dt_clf = DecisionTreeClassifier(random_state=156)
```

```python
kfold = KFold(n_splits=5,shuffle=True)
```

```python
cv_accuracy=[]
```

```python
iris.data.shape
```

```python
n_iter = 0
for train_index,test_index in kfold.split(iris.data):
    X_train,X_test = iris.data[train_index],iris.data[test_index]
    y_train,y_test = iris.target[train_index],iris.target[test_index]
    dt_clf.fit(X_train,y_train)
    pred = dt_clf.predict(X_test)
    accuracy = accuracy_score(y_test,pred)
    cv_accuracy.append(accuracy)
    n_iter +=1
    print(f'{n_iter}번째 교차검증')
    print(f'학습데이터 건수 :{X_train.shape[0]}, 검증데이터 건수:{X_test.shape[0]}')
    print(f'정확도:{accuracy}')
print(np.mean(cv_accuracy))
```

```python
kfold = KFold(n_splits=5,shuffle=True)
n_iter = 0
for train_index,test_index in kfold.split(iris.data):
    X_train,X_test = iris.data[train_index],iris.data[test_index]
    y_train,y_test = iris.target[train_index],iris.target[test_index]
    # print(type(pd.Series(y_train)))
    print('학습용데이터 분포:\n',pd.Series(y_train).value_counts())
    print('검증용데이터 분포:\n',pd.Series(y_test).value_counts())
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
skfold = StratifiedKFold(n_splits=5,shuffle=True)
n_iter = 0
for train_index,test_index in skfold.split(iris.data,iris.target):
    X_train,X_test = iris.data[train_index],iris.data[test_index]
    y_train,y_test = iris.target[train_index],iris.target[test_index]
    # print(type(pd.Series(y_train)))
    print('학습용데이터 분포:\n',pd.Series(y_train).value_counts())
    print('검증용데이터 분포:\n',pd.Series(y_test).value_counts())
```

```python
skfold = StratifiedKFold(n_splits=5,shuffle=True)
n_iter = 0
cv_accuracy=[]
for train_index,test_index in skfold.split(iris.data,iris.target):
    X_train,X_test = iris.data[train_index],iris.data[test_index]
    y_train,y_test = iris.target[train_index],iris.target[test_index]
    dt_clf.fit(X_train,y_train)
    pred = dt_clf.predict(X_test)
    accuracy = accuracy_score(y_test,pred)
    cv_accuracy.append(accuracy)
    n_iter +=1
    print(f'{n_iter}번째 교차검증')
    print(f'학습데이터 건수 :{X_train.shape[0]}, 검증데이터 건수:{X_test.shape[0]}')
    print(f'정확도:{accuracy}')
print(np.mean(cv_accuracy))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.datasets import load_iris
```

```python
iris_data = load_iris()
dt_clf = DecisionTreeClassifier(random_state=156)
```

```python
scores = cross_val_score(dt_clf,
                iris_data.data,
                iris_data.target,
                scoring='accuracy',
                cv=3)
```

```python
import numpy as np
print('교차 검증별 정확도',scores)
print('평균 교차 검증 정확도',np.mean(scores))
```

```python
from sklearn.model_selection import GridSearchCV,train_test_split
import pandas as pd
```

```python
grid_param = {'max_depth':[1,2,3],
              'min_samples_split':[2,3]}
```

```python
iris = load_iris()
X_train,X_test,y_train,y_test = train_test_split(iris.data,
                                                 iris.target,
                                                 test_size=0.2,
                                                 random_state=121)
dtree = DecisionTreeClassifier()
grid_dtree = GridSearchCV(dtree,
                          param_grid=grid_param,
                          cv=3,
                          refit=True)
grid_dtree.fit(X_train,y_train)
```

```python
grid_dtree.cv_results_
```

```python
scores_df = pd.DataFrame(grid_dtree.cv_results_)
```

```python
scores_df.columns
```

```python
scores_df[['params',
           'mean_test_score',
           'rank_test_score',
           'split0_test_score',
           'split1_test_score',
           'split2_test_score' ]]
```

```python
grid_dtree.best_score_
```

```python
grid_dtree.best_index_
```

```python
pred = grid_dtree.best_estimator_.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('best estimator 정확도', accuracy_score(y_test,pred))
```
